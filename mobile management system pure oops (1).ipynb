```python
#BLL
class Mobile:
    mobile_list=[]
    def __init__(self):        #Constructor
        self.model=""
        self.brand=""
        self.prize=0.000
        self.RAM=0
        self.internal_storage=0
        self.Battery_capacity=0
        self.OS=""
        self.Network_type=0
        self.screen_size=0
        self.sim_type="Single"
        self.primary_camera=0
        self.secondary_camera=0
    
    def add_mobile(self):
        Mobile.mobile_list.append(self)
        return "Mobile added succesfully"
        
    def search_mobile(self,brand):
        for e in Mobile.mobile_list:
            if e.brand==brand:
                self.model=e.model
                self.brand=e.brand
                self.prize=e.prize
                self.RAM=e.RAM
                self.internal_storage=e.internal_storage
                self.Battery_capacity=e.Battery_capacity
                self.OS=e.OS
                self.Network_type=e.Network_type
                self.screen_size=e.screen_size
                self.sim_type=e.sim_type
                self.primary_camera=e.primary_camera
                self.secondary_camera=e.secondary_camera
        return
    def delete_mobile(self,model):
        for e in Mobile.mobile_list:
            if e.model==model:
                Mobile.mobile_list.remove(e)
                return "Mobile removed from the list"
                                 
    def modify(self,model):
         for e in Mobile.mobile_list:
            if e.model==model:
                e.model=self.model
                e.brand=self.brand
                e.prize=self.prize
                e.RAM=self.RAM
                e.internal_storage=self.internal_storage
                e.Battery_capacity=self.Battery_capacity
                e.OS=self.OS
                e.Network_type=self.Network_type
                e.screen_size=self.screen_size
                e.sim_type=self.sim_type
                e.primary_camera=self.primary_camera
                e.secondary_camera=self.secondary_camera
            return "Mobile details modified succesfully"
                                 
        
    @staticmethod
    def show_all():
        str_m="Model\tBrand\tPrize\tRAM\tStorage\tBattery\tSystem\tNetwork\tScreen Size\tsim type\tprimarycamera\tsecondarycamera"
        for e in Mobile.mobile_list:
            str_m+="{0}\t{1}\t{2}\t{3}GB\t{4}GB\t{5}mAh\t{6}\t{7}\t{8}G\t{9}inch\t{10}MP\t{11}MP".format(e.model,e.brand,e.prize,
          e.RAM,e.internal_storage,e.Battery_capacity,e.OS,e.Network_type,e.screen_size,e.sim_type,e.primary_camera,                                                    
            e.secondary_camera)
            str_m+="\n"
        return str_m
    @staticmethod
    def sort_mobile():
        Mobile.mobile_list=sorted(Mobile.mobile_list,key=lambda p:p.prize,reverse=True)
        str_m="Model\tBrand\tPrize\tRAM\tStorage\tBattery\tSystem\tNetwork\tScreen Size\tsim type\tprimarycamera\tsecondarycamera"
        for e in Mobile.mobile_list:
            str_m+="{0}\t{1}\t{2}\t{3}GB\t{4}GB\t{5}mAh\t{6}\t{7}\t{8}G\t{9}inch\t{10}MP\t{11}MP".format(e.model,e.brand,e.prize,
          e.RAM,e.internal_storage,e.Battery_capacity,e.OS,e.Network_type,e.screen_size,e.sim_type,e.primary_camera,                                                    
            e.secondary_camera)
            str_m+="\n"
        return str_m

    

        
        #PL
while(True):
    print("1.Add\n2.Search\n3.Delete\n4.modify\n5.showall\n6.sort\n0.Exit")
    ch=input("Enter your choice")
    if ch=='1':
        mob=Mobile()
        mob.model=input("Enter Mobile Model:")
        mob.brand=input("Enter Mobile Brand:")
        mob.prize=float(input("Enter Mobile Prize:"))
        mob.RAM=int(input("Enter Mobile RAM in GB:"))
        mob.internal_storage=input("Enter Mobile internal_storage in GB:")
        mob.Battery_capacity=input("Enter Mobile Battery_capacity in mAh:")
        mob.OS=input("Enter Mobile Operating System:")
        mob.Network_type=int(input("Enter Mobile Network type:"))
        mob.screen_size=int(input("Enter Mobile screen_size :"))
        mob.sim_type=input("Enter Mobile sim_type:")
        mob.primary_camera=int(input("Enter Mobile primary_camera :"))
        mob.secondary_camera=int(input("Enter Mobile secondary_camera :"))
        x=mob.add_mobile()
        print(x)
    elif ch=="2":
        mob=Mobile()
        brand=input("Enter Mobile Brand:")
        mob.search_mobile(brand)
        str_m="\tModel{0}\tBrand{1}\tPrize{2}\tRAM{3}\tInternal_Storage{4}\tBattery_capcity{5}\tSystem{6}\
        \tNetwork type{7}\tsize{8}\
        \tsim type{9}\tprimary camera{10}\tsecondary camera{11}".format(mob.model,mob.brand,mob.prize,mob.RAM,
        mob.internal_storage,mob.Battery_capacity,mob.OS,mob.Network_type,mob.screen_size,mob.sim_type,mob.primary_camera,
        mob.secondary_camera)
        print(str_m)
    elif ch=="3":
        mob=Mobile()
        model=input("Enter Mobile Model:")
        mob.delete_mobile(model)
    elif ch=="4":
        mob=Mobile()
        model=input("Enter Mobile Model to modify:")
        mob.model=model
        mob.brand=input("Enter Mobile Brand:")
        mob.prize=float(input("Enter Mobile Prize:"))
        mob.RAM=int(input("Enter Mobile RAM in GB:"))
        mob.internal_storage=input("Enter Mobile internal_storage in GB:")
        mob.Battery_capacity=input("Enter Mobile Battery_capacity in mAh:")
        mob.OS=input("Enter Mobile Operating System:")
        mob.Network_type=int(input("Enter Mobile Network type:"))
        mob.screen_size=int(input("Enter Mobile screen_size :"))
        mob.sim_type=input("Enter Mobile sim_type:")
        mob.primary_camera=int(input("Enter Mobile primary_camera :"))
        mob.secondary_camera=int(input("Enter Mobile secondary_camera :"))
        mob.modify(model)
        
    elif ch=="5":
        x=Mobile.show_all()
        print(x)
    elif ch=="6":
        x=Mobile.sort_mobile()
        print(x)
          
            
            
            
        
    
        
             
        
        
        
        
        
    
        
        
```
