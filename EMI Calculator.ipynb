```python
#EMI calculator
#Function emi system for banks loan scheme
#3 type bank 1)SBI 2)HDFC 3)HSBC
#3 tyoe of loan scheme-home,vehicle,personal
#operation EMI for particular loan amount per month
#EMI-time year --convert into month 
#rate=%percentage
#if the  interest rate per annum is 11%
#then the rate of interest will be 11/(12*100)]
#EMI=[P*R*(1+R)^N]/[(1+R)^N-1]
#P=amount of loan
#R-rare of interest 
#N-time of tenure
```

```python
def emi_cal(P,r,t):
    R=r/(12*100)    #one month interest
    T=t*12
    emi=((P*R*((1+R)**T))/(((1+R)**T)-1))
    return emi
        
while(True):
    print("____________LOAN EMI SYSTEM______________")
    print()
    print("\n1:SBI\n2:HDFC\n3:HSBC\n4:Exit")
    print("Select your bank from the choice---:")
    sb=int(input("Enter Bank Choice:"))
    if(sb==1):
        print("Welcome To SBI Loan Scheme")
        print()
        print("1:Home 2:Vehicle 3:Personal")
         #bank SBI home loan -7.98%
         #Personal-8.56%
         #vehicle-9.78%
        ch=int(input("Enter Choice for Loan:"))
        if ch==1:
            p=float(input("Enter loan Amount:"))
            t=int(input("Enter Loan Tenure in year:"))
            r=7.88
            print("Montally EMI will be:",emi_cal(p,r,t))
        elif ch==2:
            p=float(input("Enter loan Amount:"))
            t=int(input("Enter Loan Tenure in year:"))
            r=8.56
            print("Montally EMI will be:",emi_cal(p,r,t))
        elif ch==3:
            p=float(input("Enter loan Amount:"))
            t=int(input("Enter Loan Tenure in year:"))
            r=8.78
            print("Montally EMI will be:",emi_cal(p,r,t))
        else:
             break
      
    elif(sb==2):
        print("Welcome To HDFC Loan Scheme")
        print()
        p=float(input("Enter loan Amount:"))
        t=int(input("Enter Loan Tenure in year:"))
        print("1:Home 2:Vehicle 3:Personal")
         #bank HDFC home loan -8.98%
         #Personal-8.06%
         #vehicle-9.88%
        c2=int(input("Enter Choice for Loan:"))
        if c2==1:
            r=8.98
            print("Montally EMI will be:",emi_cal(p,r,t))
        elif c2==2:
            r=8.06
            print("Montally EMI will be:",emi_cal(p,r,t))
        elif c2==3:
            r=9.88
            print("Montally EMI will be:",emi_cal(p,r,t))
        elif c2==4:
             break
    elif(sb==3):
         while(True):
                print("Welcome To HSBC Loan Scheme")
                print()
                loan_data={"home":7.65,"Vehicle":9.50,"personal":8.99}
                p=float(input("Enter loan Amount:"))
                t=int(input("Enter Loan Tenure in year:"))
                print("1:Home 2:Vehicle 3:Personal")
         
                c3=int(input("Enter Choice for Loan:"))
                if c3==1:
                        r=loan_data['home']
                        print("Montally EMI will be:",emi_cal(p,r,t))
                elif c3==2:
                         r=loan_data['Vehicle']
                         print("Montally EMI will be:",emi_cal(p,r,t))
                elif c3==3:
                         r=loan_data['personal']
                         print("Montally EMI will be:",emi_cal(p,r,t))
                elif c3==4:
                         break
    else:
         print("Invalid Choice")
         
        
```
