```python
#STUDENT MANAGEMENT SYSTEM - It maintains record of student in fiels like roll no,name,class,marks of five subjects,percentage,status
#Calculates the percentage.here the percentage function is a variable length function in order to deal with any new subject
#that can be added in future.
#On the basis of percentage it all finds out whether the student is pass or fail.


class Student:
    def __init__(self):
        self.Roll_No=0
        self.Name=""
        self.Class=""
        self.sci=0
        self.maths=0
        self.eng=0
        self.hindi=0
        self.ssc=0
        self.percent=0
        self.status=""
        
list_std=[]

def percentage(*args):
    sum=0
    if len(args)!=0:
        for e in args:
            sum=sum+e
    per=sum/len(args)
    return per

def status(percent):
    if percent>35:
        return "Pass"
    else:
        return "Fail"
    
def add(std):
    list_std.append(std)
    return "Student Added Succesfully" 
    
def search(Roll_no):
    for e in list_std:
        if e.Roll_No==Roll_no:
            return e
    return False   

def modify_student_details(no,modify):
    for e in list_std:
        if e.Roll_No==no:
                e.Name=modify.Name
                e.Class=modify.Class
                e.sci=modify.sci
                e.maths=modify.maths
                e.eng=modify.eng
                e.hindi=modify.hindi
                e.ssc=modify.ssc
                e.percent=modify.percent
                e.status=modify.status
                return "Student details modified succesfully"
    return "Roll no not found"

def Delete(no):
    for e in list_std:
        if e.Roll_No==no:
            list_std.remove(e)
            return "Student Deleted Successfully"
    return "Roll No not found"

def Show_all():
    msg=""
    for e in list_std:
            msg+="Student Roll No:"+str(e.Roll_No) +"\tStudent Name:"+e.Name+"\tStudent Class:"+str(e.Class)+\
            "\tMarks:"+"\tScience:"+str(e.sci)+"\tMaths:"+str(e.maths)+"\tEnglish:"+str(e.eng)+"\tHindi:"+str(e.hindi)+\
            "\tSSC:"+str(e.ssc)+"\tPercentage:"+str(e.percent)+"\tStatus:"+e.status+"\n"
    return msg        
            

        
    
while(True):
    
    print("----------WELCOME TO STUDENT MANAGEMENT SYSTEM-----------")
    print("\n1.Add Student\n2.Search Student\n3.Modify\n4.Delete Record\n5.Show All")
    ch=input("Enter your choice")
    if ch=='1':
        std=Student()
        std.Roll_No=int(input("Enter Roll no of the student"))
        std.Name=input("Enter Name of Student")
        std.Class=input("Enter class")
        std.sci=int(input("Enter Science Marks:"))
        std.maths=int(input("Enter Maths Marks:"))
        std.eng=int(input("Enter English Marks:"))
        std.hindi=int(input("Enter Hindi Marks:"))
        std.ssc=int(input("Enter SSC Marks:"))
        std.percent=percentage(std.sci,std.maths,std.eng,std.hindi,std.ssc)
        std.status=status(std.percent)
        res=add(std)
        print(res)
        
    elif ch=='2':
        Roll_no=int(input("Enter Student Rollno"))
        std=search(Roll_no)
        if std==False:
            print("Student Not found")
        
        elif std:
            print("Student Roll No:",std.Roll_No)
            print("Student Name:",std.Name)
            print("Student Class:",std.Class) 
            print("Marks in Final Exam out of 100")
            print("\nScience:",std.sci,"\nMaths:",std.maths,"\nEnglish",std.eng,"\nHindi",std.hindi,"\nSSC",std.ssc)
            print("Percentage:",std.percent)       
            print("Status:",std.status)
    
    elif ch=="3":
            no=int(input("Enter Roll no of the student"))
            print("Enter Modification details:")
            modify=Student()
            modify.Name=input("Enter Name of Student")
            modify.Class=input("Enter class")
            modify.sci=int(input("Enter Science Marks:"))
            modify.maths=int(input("Enter Maths Marks:"))
            modify.eng=int(input("Enter English Marks:"))
            modify.hindi=int(input("Enter Hindi Marks:"))
            modify.ssc=int(input("Enter SSC Marks:"))
            modify.percent=percentage(modify.sci,modify.maths,modify.eng,modify.hindi,modify.ssc)
            modify.status=status(modify.percent)
            msg=modify_student_details(no,modify)
            print(msg)
    
    elif ch=="4":
            no=int(input("Enter Roll no of the student you want to delete"))
            msg=Delete(no)
            print(msg)
    elif ch=="5":
            msg=Show_all()
            print(msg)
        
       
        
```
